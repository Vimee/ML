```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0, 2.0,0.01)
s = 1+np.cos(2*np.pi*t)
plt.plot(t,s,'--')
plt.grid()
#plt.plot(t,s,'g')
#plt.plot(t,s,'')
#plt.plot(t,s,'*')

plt.xlabel('Time(sec)')
plt.ylabel('Voltage(mV)')
plt.title('Cosine wanes plot(Cos(x))')
plt.show()
```
